# Quickstart manual

1. New Terminal
	- cd nome cartella

2. Installa ambiente virtuale
	- py -3.12 -m pip install virtualenv
	- py -3.12 -m virtualenv .venv

3. Attiva ambiente virtuale
	- .\.venv\Scripts\Activate.ps1

4. Installa le librerie da requirement
	- py -3.12 -m pip install -r src\requirement.txt

5. Avvia l'app
	- python -m streamlit run src\main.py
	- se non vanno import Ctrl+Shift+P Python: Select Interpreter

```python
"""CELLA DI SETUP: 
esegui questo script per creare un ambiente virtuale e installare le dipendenze.
Solo la prima volvta, poi puoi eseguire direttamente l'app Streamlit:
 - streamlit run src/main.py`.
 - eseguire la CELLA DI AVVIO
"""

from pathlib import Path
import subprocess

project_root = Path.cwd()
venv_python = project_root / ".venv" / "Scripts" / "python.exe"
requirements_path = project_root / "src" / "requirement.txt"

setup_commands = [
    ["py", "-3.12", "-m", "pip", "install", "virtualenv"],
    ["py", "-3.12", "-m", "virtualenv", ".venv"],
    [str(venv_python), "-m", "pip", "install", "-r", str(requirements_path)],
]

for command in setup_commands:
    print("$", " ".join(command))
    subprocess.run(command, cwd=project_root, check=True)

print("Ambiente pronto in", project_root / ".venv")
```

```python
"""CELLA DI AVVIO: apre un terminale ed esegue Streamlit."""

from pathlib import Path
import subprocess

project_root = Path.cwd()

subprocess.run(
    ["python", "-m", "streamlit", "run", "src\\main.py"],
    cwd=project_root,
    check=True,
)
```
